```python
import sys
sys.path.append("..")

from src.rag_pipeline import RAGPipeline
```

```python
VECTOR_STORE_PATH = "../vector_store"

rag = RAGPipeline(
    vector_store_path=VECTOR_STORE_PATH,
    top_k=5
)
```

```python
query = "Why are customers unhappy with credit cards?"

result = rag.ask(query)

print("QUESTION:")
print(result["question"])
print("\nANSWER:")
print(result["answer"])
```

```python
for i, doc in enumerate(result["sources"], 1):
    print(f"\nSource {i}")
    print("Product:", doc.metadata.get("product"))
    print("Issue:", doc.metadata.get("issue"))
    print(doc.page_content[:400], "...")
```

```python
evaluation_questions = [
    "Why are people unhappy with credit cards?",
    "What are common issues with money transfers?",
    "What complaints are frequent for personal loans?",
    "Are customers reporting fraud-related issues?",
    "What delays do customers experience with savings accounts?",
]
```

```python
evaluation_results = []

for q in evaluation_questions:
    result = rag.ask(q)
    evaluation_results.append({
        "question": q,
        "answer": result["answer"],
        "sources": result["sources"][:2]
    })
```

```python
from IPython.display import Markdown, display

table_md = "| Question | Generated Answer | Quality (1â€“5) | Comments |\n"
table_md += "|---|---|---|---|\n"

for r in evaluation_results:
    table_md += f"| {r['question']} | {r['answer'][:120]}... |  |  |\n"

display(Markdown(table_md))
```
